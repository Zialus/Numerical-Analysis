```python
def bisect_iterative(function, a, b, eps):
    m = a
    vfa = function(a)
    error_iter = abs(b - a)
    iteration = 0
    while error_iter > eps:
        m = (a + b) / 2
        vfm = function(m)
        if vfm == 0:
            error_iter = 0
        elif vfm * vfa < 0:
            b = m
        else:  # (vfm*vfa > 0)
            a = m

        error_iter /= 2
        iteration += 1
        print("error_iter: ", error_iter)

    return m, error_iter, iteration
```

```python
def simple_iterative(function, x0, eps, nmax):
    x1 = function(x0)
    error_iter = abs(x1 - x0)
    i = 1
    while (error_iter > eps) and (i <= nmax):
        x0 = x1
        x1 = function(x0)
        error_iter = abs(x1 - x0)
        i += 1

    if i > nmax:
        print(
            "não foi possível ao fim de ",
            nmax,
            " iterações encontrar a solução com o erro pretendido",
        )
    else:
        print(x1, error_iter)
```

```python
import math

def example_function_a(x):
    return (0.123**x)-x

bisect_iterative(example_function_a,0,1,5*10**-4)
```

```python
def example_function_b(x):
    return x**3 - 2*((math.e)**-x)

bisect_iterative(example_function_b,0,1,5*10**-5)
```
