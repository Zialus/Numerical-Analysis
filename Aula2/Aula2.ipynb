```python
import math
```

```python
x = 0.099
X = 0.1

deltaX = X - x

absolute_err = deltaX
relative_err = deltaX/X

print("X:")
print(absolute_err)
print(relative_err)

y = 0.011
Y = 0.01

deltaY = Y - y

absolute_err = deltaY
relative_err = deltaY/Y

print("Y:")
print(absolute_err)
print(relative_err)
```

```python
def f(x):
    return x * (math.sqrt(x+1) - math.sqrt(x))

for x in [10**i for i in range(0,40)]:
    fx = f(x)
    print(f'{x:40} ==> {fx}')
```

```python
sqrt_of_3 = 1.732

exact = 1 / (2 + math.sqrt(3))**4
approx = 1 / (2 + sqrt_of_3)**4

print("1:")
print(exact)
print(approx)

exact = 97 - 56 * math.sqrt(3)
approx = 97 - 56 * sqrt_of_3

print("2:")
print(exact)
print(approx)
```
