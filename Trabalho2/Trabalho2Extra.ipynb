```python
import matplotlib.pyplot as plt

import numpy as np

def fixedp(f,x0,eps=5e-10,maxiter=1000):
    """ Fixed point algorithm """
    err = 1
    itr = 1
    xp = []
    while(err > eps and itr <= maxiter):
        x1 = f(x0)       # fixed point equation
        err = abs(x0-x1) # error at the current step
        x0 = x1
        xp.append(x0)    # save the solution of the current step
        itr += 1
    return x1,xp

f = lambda x: np.float64(x)**2-np.sin(np.float64(x)+0.15)
```

```python
%matplotlib inline

x_start = -0.1
xf,xp = fixedp(f,x_start)

x = np.linspace(-1,2,1000)
y = f(x)

fig1, ax1 = plt.subplots(figsize=(15, 15))
ax1.grid(True, which='both')

ax1.plot(x,y,xp,f(xp),'bo')
ax1.plot(x_start,f(x_start),'ro')
ax1.plot(xf,f(xf),'go')
ax1.plot(x,x,'k')

fig1.show()
```

```python
up_to = 1000

for elem in xp[0:up_to]:
    print(elem)
```

```python
%matplotlib inline

fig2, ax2 = plt.subplots(figsize=(15, 15))
ax2.grid(True, which='both')

ax2.plot(range(up_to),xp[:up_to],'bo')

fig2.show()
```
