```python
def newton(f,Df,x0,epsilon,max_iter):
    xn = x0
    for n in range(0,max_iter):
        fxn = f(xn)
        if abs(fxn) < epsilon:
            print('Found solution after',n,'iterations.')
            return xn
        Dfxn = Df(xn)
        if Dfxn == 0:
            print('Zero derivative. No solution found.')
            return None
        xn = xn - fxn/Dfxn
    print('Exceeded maximum iterations. No solution found.')
    return None
```

```python
def our_function(x):
    return x**2 - x - np.sin(x + 0.15)

def derivative_of_our_function(x):
    return 2*x - 1 - np.cos(x+0.15)

def second_derivative_of_our_function(x):
    return 2 + np.sin(x+0.15)

def function_for_iterative_method(x):
    return x**2-np.sin(x+0.15)
```
