```python
def simple_iterative(function, x0, eps, max_iter):
    '''
    f: fuction for which we want to find a root
    x0: initial point
    eps: erro maximo de aproximacao
    '''
    x1 = function(x0)
    error_iter = abs(x1 - x0)
    i = 1
    while (error_iter > eps) and (i <= max_iter):
        x0 = x1
        x1 = function(x0)
        error_iter = abs(x1 - x0)
        i += 1
#         print(x0,x1,error_iter)
    if i > max_iter:
        print('não foi possível ao fim de', max_iter, 'iterações encontrar a solução com o erro pretendido')
    else:
        print(x1, erroiter)
```

```python
def newton(f, Df, x0, eps, max_iter):
    '''
    f: funcao para a qual estamos a procura da raiz
    Df: derivada da funcao
    x0: ponto de arranque
    eps: erro maximo da aproximacao
    max_iter: numero maximo de iteracoes
    '''
    xn = x0
    for n in range(0,max_iter):
        fxn = f(xn)
        Dfxn = Df(xn)
        if Dfxn == 0:
            print('Derivada com valor 0. Solucao nao encontrada.')
            return None
        old_xn = xn
        xn = xn - fxn/Dfxn
        if abs(old_xn - xn) < eps:
            print('Encontrada uma solucao ao fim de',n,'iteracoes.')
            return xn
    print('Excedido limite de iteracoes. Solucao nao encontrada.')
    return None
```

```python
def our_function(x):
    return x**2 - x - np.sin(x + 0.15)

def derivative_of_our_function(x):
    return 2*x - 1 - np.cos(x+0.15)

def second_derivative_of_our_function(x):
    return 2 + np.sin(x+0.15)

def function_for_iterative_method(x):
    return x**2-np.sin(x+0.15)
```
